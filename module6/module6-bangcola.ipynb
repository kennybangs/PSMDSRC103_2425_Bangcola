```python
import random
random.seed(0)

salaries = [round(random.random() * 1000000, -3) for _ in range(100)]
salaries
```

```python
#mean
a = 0
#get sum of observations
for x in salaries:
    a = a +x
#divide sum by number of observations
meanSalary = (a/len(salaries))
print(meanSalary)
```

```python
#median
#sort list ascending
salaries.sort()
#find middle of list
midpoint = int(len(salaries)/2)-1
#since it is even, get the average of 50th and 51st observation
print((salaries[midpoint]+salaries[midpoint+1])/2)
```

```python
#mode
#import collections first
import collections
#initialize instance of Counter on salaries
counted =  collections.Counter(salaries)
#get max value of the values of the dict
maxValue = max(counted.values())
#print(maxValue)
#search dictionary using the key
for x in counted:
    if counted[x] == maxValue:
        print(x)
    else:
        pass

    
```

```python
#sample variance
#first get square of the variances of each observation
variance = []
for x in salaries:
    variance.append((x-meanSalary)**2)
#sum the square of the variances and divide by length - 1
sampleVariance = sum(variance)/len(salaries)-1
print(sampleVariance)
```

```python
#sample standard deviation
#this is the sqrt of the sample variance

import math

sampleStdev = math.sqrt(sampleVariance)
print(sampleStdev)
```

```python
#Comparison with statistics module
import statistics as stat
#mean - our value: 585690.0
print(stat.mean(salaries))
#CORRECT: mean() and our guess match

#median - our value: 589000.0
print(stat.median(salaries))
#CORRECT: median() and our guess match

#mode - our value: 477000.0
print(stat.mode(salaries))
#CORRECT: mode() and our guess match

#sample variance
print(stat.variance(salaries))
#not matching... why?

#sample stdev
print(stat.stdev(salaries))
#not matching... why?
```

```python
#Exercise 2

#Range = largest - smallest value
range = max(salaries)-min(salaries)
print(range)

#coefficient of variation = stdev/mean * 100
covar = (stat.stdev(salaries)/stat.mean(salaries))*100
print(covar)

#IQR = difference between Q3 and Q1
#find index of q1
indexQ1 = int((25/100)*len(salaries))
#print(indexQ1)
q1 = salaries[indexQ1]
#find index of q3
indexQ3 = int((75/100)*len(salaries))
#print(indexQ3)
q3  = salaries[indexQ3]
#get iqr
iqr = q3 - q1
print(iqr)

#quartile coefficient of dispersion = (q3-q1)/(q3+q1)
qcod = (q3-q1)/(q3+q1)
print(qcod)
```
