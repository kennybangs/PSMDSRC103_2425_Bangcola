```python
#Part 1

import random
import math
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]

#Without using the statistics module, compute for the ff:
#Mean
#Median
#Mode
#Sample variance
#Sample stdev

length = len(salaries)

#Mean

mean = sum(salaries)/length
print("Mean: " + str(mean))

#Median
if len(salaries) % 2 == 0:
    median = (salaries[int(length/2)]+salaries[int(length/2)+1])/2
else:
    median = salaries[int((length+1)/2)]

print("Median: " + str(median))

#Mode
countDict = {}
#count number of occurences per list item
for x in salaries:
    countDict[x] = salaries.count(x)

ranked = list(countDict.values())
ranked.sort(reverse=True)
ranked[0]

for value, occurences in countDict.items():  # for name, age in dictionary.iteritems():  (for Python 2.x)
    if occurences == ranked[0]:
        print("Mode: " + str(value))

#Sample variance
sampleSums = []
for x in salaries:
    sampleSums.append(x-mean)
#print(sampleSums)
sampleVariance = sum(sampleSums)**2/(length-1)
print("Sample variance: " + str(sampleVariance))

#Sample stdev
sampleStdev = math.sqrt(sampleVariance)
print(sampleStdev)
```

```python
#using statistics module

import statistics as stat

print("Mean: " + str(stat.mean(salaries)))
print("Median: " + str(stat.median(salaries)))
print("Mode: " + str(stat.mode(salaries)))
print("Variance: " + str(stat.variance(salaries)))
print("Stdev: " + str(stat.stdev(salaries)))
```

```python
#Change the outcome column to Diagnosis

diabetes_df.rename(columns={"Outcome":"Diagnosis"},inplace=True)
diabetes_df.columns
```

```python
#Create a new column classification

classification=[]

for x in diabetes_df['Diagnosis']:
    if x == 1:
        classification.append("Diabetes")
    else:
        classification.append("No diabates")

diabetes_df.insert(9,'Classification',classification)
```
